```python
CREATE DATABASE air_pollution;

USE air_pollution;
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('air_pollution.db')
```

```python
df.to_sql(
'pollution_data',
conn,
if_exists='replace',
index=False
)
```

```python
query = '''

SELECT city,
AVG(pollutant_avg) AS avg_pollution

FROM pollution_data

GROUP BY city

ORDER BY avg_pollution DESC

LIMIT 10

'''

result = pd.read_sql(query, conn)

result
```
